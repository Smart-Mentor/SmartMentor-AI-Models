```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# ================================
# LOAD DATA
# ================================
df = pd.read_csv("Data.csv")

print("Data loaded")

# ================================
# VECTORIZE JOBS
# ================================
vectorizer = TfidfVectorizer()

job_vectors = vectorizer.fit_transform(df["cleaned_skills_text_v2"])

print("Jobs vectorized")

# ================================
# CV INPUT
# ================================
def recommend_jobs(cv_text, top_n=3):
    # vectorize CV
    cv_vector = vectorizer.transform([cv_text])

    # similarity
    similarities = cosine_similarity(cv_vector, job_vectors)

    # get top jobs
    top_indices = similarities[0].argsort()[-top_n:][::-1]

    results = df.iloc[top_indices][[
        "Job Title",
        "Company Name",
        "cleaned_skills_text_v2"
    ]]

    results["score"] = similarities[0][top_indices]

    return results


# ================================
# TEST
# ================================
cv_example = "python machine learning data analysis pandas numpy"

results = recommend_jobs(cv_example)

print(results)
```

```python
import pickle

# save vectorizer
with open("vectorizer.pkl", "wb") as f:
    pickle.dump(vectorizer, f)

# save job vectors
with open("job_vectors.pkl", "wb") as f:
    pickle.dump(job_vectors, f)

# save dataframe
df.to_pickle("data.pkl")

print("Model saved!")
```
